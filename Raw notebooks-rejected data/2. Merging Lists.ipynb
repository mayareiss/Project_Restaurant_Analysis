Merging data with Bonnie's list, provided from https://data.sfgov.org/Health-and-Social-Services/Restaurant-Scores-LIVES-Standard/pyih-qa8i

```python
import pandas as pd
maya = pd.read_csv("SF_Registered_Food_Businesses.csv")
maya.head()
```

```python
bonnie = pd.read_csv("Restaurant_Score_Output.csv")
bonnie.head()
```

```python
#since the column titles aren't identical, renaming for the purpose of merging

maya.rename(columns = {'DBA Name':'business_name'}, inplace = True)
maya.head()
```

```python
merged = pd.merge(maya, bonnie, on="business_name", how="outer")
merged
```

```python
#cleaning data to remove the non-matches

clean = merged.dropna(subset=['business_id'])
clean
```

```python
clean = clean.dropna(subset=['Location Id'])
clean
```

```python
clean.columns
```

```python
#removing extraneous columns
open_sf_restaurants = clean[[
                        'business_id', 'business_name', 'Street Address', 'City', 'State', 'Source Zipcode',
                        'Business Start Date', 'Business End Date', 'Location Start Date',
                        'Location End Date', 'LIC Code', 'LIC Code Description', 
                        'Supervisor District','Neighborhoods - Analysis Boundaries']]

open_sf_restaurants
```

```python
open_sf_restaurants.to_csv('./Open_SF_Restaurants.csv', encoding="utf-8", index=False, header=True)
```

In the end, decided not to use this data as the matches aren't accurate. One file has a Business ID, and another file had a location ID, but they were not interchangeable. We matches based on Business Name (DBA Name) and added the Business ID, but as some restaurants had multiple locations, we cannot say for sure that this is accurate. But this was a good learning experience
