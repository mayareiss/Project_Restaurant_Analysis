```python
import pandas as pd
import pandas as pd
import numpy as np
import requests
import time
import json
import os

from key import ykey

#API URL
headers = {'Authorization': 'Bearer %s' % ykey}

#pulling in the SF data of Registered Food Businesses
data = pd.read_csv("Cleaned_Categories.csv")
data
```

```python
neighb_data = data[data['Neighborhood'].isnull()]
neighb_data
                   
```

```python
lat_long = neighb_data[['DBA Name', 'Street Address', 'Yelp Name',
                        'Price Level', 'Rating', 'Review Count', 'Top Yelp Category', 'Is Closed',
                        'Yelp Address', 'Zipcode',
                        'Neighborhood', 'Yelp ID']]
```

```python
lat_long['Latitude']=""
lat_long['Longitude']=""
lat_long
```

```python
# use iterrows to iterate through pandas dataframe
for index, row in lat_long.iloc[1:].iterrows():

    # get Yelp ID from df
    term = row['Yelp Name']
    location = row ['Street Address']
    url='https://api.yelp.com/v3/businesses/search'

    # assemble url and make API request
    params = params={'term':term,"location": location}
    req = requests.get(url, headers=headers, params=params)
    response = req.json()
    # extract results
    results = response
    
    try:
        lat_long.loc[index, 'Latitude'] = results['businesses'][0]['coordinates']['latitude']
    except (KeyError, IndexError):
        lat_long.loc[index, 'Latitude'] = "No record in Yelp"
    
    try:
        lat_long.loc[index, 'Longitude'] = results['businesses'][0]['coordinates']['longitude']
    except (KeyError, IndexError):
        lat_long.loc[index, 'Longitude'] = "No record in Yelp"
        
```
